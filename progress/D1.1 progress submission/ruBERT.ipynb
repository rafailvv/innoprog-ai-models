```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM

def generate_text(prompt, model_name="DeepPavlov/rubert-base-cased", max_length=50):
    # Загрузка токенизатора и модели Rubert
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModelForCausalLM.from_pretrained(model_name)

    # Преобразование текста в токены
    inputs = tokenizer.encode(prompt, return_tensors="pt")

    # Генерация текста с помощью модели
    outputs = model.generate(inputs, max_length=max_length, num_return_sequences=1, pad_token_id=tokenizer.eos_token_id)

    # Декодирование сгенерированного текста
    generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

    return generated_text

# Пример использования
prompt = "Пользователь вводит в консоль количество звезд"
response = generate_text(prompt)
print(response)
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM

def generate_text(prompt, model_name="gpt2", max_length=50):
    # Загрузка токенизатора и модели GPT-2
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModelForCausalLM.from_pretrained(model_name)

    # Преобразование текста в токены
    inputs = tokenizer.encode(prompt, return_tensors="pt")

    # Генерация текста с помощью модели
    outputs = model.generate(inputs, max_length=max_length, num_return_sequences=1, pad_token_id=tokenizer.eos_token_id)

    # Декодирование сгенерированного текста
    generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

    return generated_text

# Пример использования
prompt = "Пользователь вводит в консоль количество звезд"
response = generate_text(prompt)
print(response)
```

```python
from transformers import BertForMaskedLM,BertTokenizer, pipeline
model=BertForMaskedLM.from_pretrained('sberbank-ai/ruBert-base')
tokenizer=BertTokenizer.from_pretrained('sberbank-ai/ruBert-base', do_lower_case=False)
unmasker = pipeline('fill-mask', model=model,tokenizer=tokenizer)
```

```python
text = "Пользователь вводит в консоль количество звезд в [MASK] данные"
print([i['sequence'] for i in unmasker(text)])
```

```python

from random import choice
text = "Пользователь вводит в консоль количество звезд"
for i in range(5):
  res = [i['sequence'] for i in unmasker(text + " [MASK]")]
  print(res)
  text = choice(res)
print(text)
```
