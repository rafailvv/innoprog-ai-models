```python
import json

# Opening JSON file
liza_f = open('datasets/chat/raw/–ª–∏–∑–∞.json')

data = json.load(liza_f)
data['messages']
```

```python
messages = []

for msg in data['messages']:
    temp_dict = {'id': msg['id'], 'from': msg['from_id']}
    txt = ''
    for t in msg['text_entities']:
        txt += t['text']
    temp_dict['text'] = txt
    temp_dict['date'] = msg['date_unixtime']
    messages.append(temp_dict.copy())

messages
```
